```python
# 1.1 Finding missing values in Meter B
dataframe = dataframe.replace('?',np.NaN)

print('Number of instances = %d' % (dataframe.shape[0]))
print('Number of attributes = %d' % (dataframe.shape[1]))

print('Number of missing values:')
for col in dataframe.columns:
    print('\t%s: %d' % (col,dataframe[col].isna().sum()))
```

```python
# 1.2 Finding duplicate rows in Meter B
sum(dataframe.duplicated())
```

```python
# 1.3 Finding outliers in Meter B with Boxplots
dataframe.plot(kind="box", subplots=True, layout=(20,3), figsize=(20,60))
```

```python
# 1.3.1 Determine if it is a normal distribution or a skewed distribution with Histrogram.
# If it is a normal distribution, use z-score to remove the outliers.
# Else, using IQR to remove the outliers.
import matplotlib.colors as mcolors
colors = list(mcolors.CSS4_COLORS.keys())[10:]
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor=colors[i])
        ax.set_title(feature+" Histogram",color=colors[35])
        ax.set_yscale('log')
    fig.tight_layout() 
    plt.savefig('Histograms.png')
    plt.show()
draw_histograms(dataframe,dataframe.columns,9,6)
```

**<font color='red'>Conclusion</font>**<br>
It is a skewed distribution, therefore IQR will be used to remove the outliers.

```python
# define a function called outliers which returns a list of index of outliers
def outliers(df,ft):
    Q1=df[ft].quantile(0.25)
    Q3=df[ft].quantile(0.75)
    IQR=Q3-Q1

    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR

    ls = df.index[(df[ft]<lower_bound )|(df[ft]>upper_bound)]
    return ls
```

```python
#extracting outliers of all the columns
index_list=[]
dataframe2 = dataframe.iloc[:,dataframe.columns != 'Class']# Drop the class attribute in the data using drop() function as class attribute is qualitative and not quantitative
for features in dataframe2.columns:
    index_list.extend(outliers(dataframe2,features))
```

```python
# define a function called remove, which returns a cleaned dataframe without outliers
def remove(df,ls):
    #sort and unique the index list
    ls=sorted(set(ls))

    #remove the rows based on the index list
    df=df.drop(ls)

    return df 
```

```python
# 1.4 Removing the outliers and getting the size of the new dataframe
dataframe2 = pd.concat([dataframe2, dataframe[['Class']]], axis = 1)
```

```python
dataframe_cleaned=remove(dataframe2,index_list)
dataframe_cleaned.shape
```

```python
# See if the outliers has been remved
dataframe_cleaned.plot(kind="box", subplots=True, layout=(20,3), figsize=(20,60))
```

#### 2. Data Transformation (Feature Scaling)


Before doing feature scaling, we need to find out what is the best number for our dataset

```python
new_X = dataframe_cleaned.iloc[:,dataframe_cleaned.columns != 'Class']
new_y = dataframe_cleaned.Class

print(new_X)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(new_X)
```

#### 3. Data Reduction (PCA)

```python
from sklearn.decomposition import PCA
PCA = PCA(n_components=0.95)# with 95% accuracy
components = PCA.fit_transform(X)
PCA.components_
```

```python
cumVar = pd.DataFrame(np.cumsum(PCA.explained_variance_ratio_)*100, 
                      columns=["cumVarPerc"])
expVar = pd.DataFrame(PCA.explained_variance_ratio_*100, columns=["VarPerc"])
pd.concat([expVar, cumVar], axis=1)
```

**<font color='red'>Conclusion</font>**<br>
I only need 4 pca to have 95% accuracy.

```python
# Do feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(new_X)

# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)

# Get the transformed dataset
X_pca = pd.DataFrame(X_pca)
print(X_pca.head())
print("\nSize: ")
print(X_pca.shape)

# Calculate the variance explained by priciple components
print('Variance of each component:', pca.explained_variance_ratio_)
print('\nTotal Variance Explained:', round(sum(list(pca.explained_variance_ratio_))*100, 2))
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sns

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, new_y, test_size=0.20, 
                                                            random_state=42)

# Initialize the logistic regression model
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=2500)

# Train the model
clf.fit(X_train_pca, y_train)

# Make predictions
y_pred = clf.predict(X_test_pca) # Predictions
y_true = y_test # True values

# Measure accuracy
from sklearn.metrics import accuracy_score
import numpy as np
print("Train accuracy:", np.round(accuracy_score(y_train, 
                                                 clf.predict(X_train_pca)), 2))
print("Test accuracy:", np.round(accuracy_score(y_true, y_pred), 2))

# Make the confusion matrix
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_true, y_pred)
score = clf.score(X_test_pca, y_test)
plt.figure(figsize=(9,9))
sns.heatmap(cf_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred,zero_division=1))
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
kf = KFold(n_splits=4)
mae_train = []
mae_test = []
for train_index, test_index in kf.split(X):

   model = KNeighborsClassifier(n_neighbors=2)
   model.fit(X_train_pca, y_train)
   y_train_pred = model.predict(X_train_pca)
   y_test_pred = model.predict(X_test_pca)
   mae_train.append(mean_absolute_error(y_train, y_train_pred))
   mae_test.append(mean_absolute_error(y_test, y_test_pred))
```

```python
folds = range(1, kf.get_n_splits() + 1)
plt.plot(folds, mae_train, 'o-', color='green', label='train')
plt.plot(folds, mae_test, 'o-', color='red', label='test')
plt.legend()
plt.grid()
plt.xlabel('Number of fold')
plt.ylabel('Mean Absolute Error')
plt.show()
```

**<font color='red'>Conclusion</font>**<br>

1. Precision is a measure of how many of the positive predictions made are correct.The model was able to have 100% positive predicitions.
2. Recall is a measure of how many of the positive cases the classifier correctly predicted, over all the positive cases in the data. The model was abke to 100% classify correctly.
3. F1-Score is a measure of a test's accuracy. It is calculated from the precision and recall of the test, where the precision is the number of true positive results divided by the number of all positive results, including those not identified correctly, and the recall is the number of true positive results divided by the number of all samples that should have been identified as positive. 
Therefore, with the model have f1-score of 100%, it means it has perfect precision and recall.
