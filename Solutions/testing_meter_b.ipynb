```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

dataframe = pd.read_csv("Meter B", sep='\t',header=None)
dataframe.to_csv('Meter_B.csv',header=None)
dataframe.columns=['Profile factor','Symmetry','Crossflow','Swirl angle','Flow velocity 1','Flow velocity 2','Flow velocity 3','Flow velocity 4',
'Average flow', 'Speed of Sound 1','Speed of Sound 2','Speed of Sound 3','Speed of Sound 4','Average Speed','Signal Strength 1','Signal Strength 2','Signal Strength 3','Signal Strength 4','Signal Strength 5',
'Signal Strength 6','Signal Strength 7','Signal Strength 8', 'Turbulence 1','Turbulence 2','Turbulence 3','Turbulence 4','Meter performance','Signal Quality 1','Signal Quality 2','Signal Quality 3','Signal Quality 4','Signal Quality 5','Signal Quality 6','Signal Quality 7','Signal Quality 8',
'Gains 1','Gains 2','Gains 3','Gains 4','Gains 5','Gains 6','Gains 7','Gains 8',
'Transit Time 1','Transit Time 2','Transit Time 3','Transit Time 4','Transit Time 5','Transit Time 6','Transit Time 7','Transit Time 8',
'Class']


print('Number of instances = %d' % (dataframe.shape[0]))
print('Number of attributes = %d' % (dataframe.shape[1]))
dataframe
```

```python

dataframe = dataframe.replace('?',np.NaN)

print('Number of instances = %d' % (dataframe.shape[0]))
print('Number of attributes = %d' % (dataframe.shape[1]))

print('Number of missing values:')
for col in dataframe.columns:
    print('\t%s: %d' % (col,dataframe[col].isna().sum()))
```

Finding the outliers among the dataset of Meter B

```python
# calculate Q1 and Q3
Q1 = dataframe.quantile(0.25)
Q3 = dataframe.quantile(0.75)

# calculate the IQR
IQR = Q3 - Q1

# filter the dataset with the IQR
IQR_outliers = dataframe[((dataframe < (Q1 - 1.5 * IQR)) |(dataframe > (Q3 + 1.5 * IQR))).any(axis=1)]
IQR_outliers
```

```python
#remove outliers
dataframe = dataframe[~((dataframe < (Q1 - 1.5 * IQR)) |(dataframe > (Q3 + 1.5 * IQR))).any(axis=1)]

print('Number of instances = %d' % (dataframe.shape[0]))
print('Number of attributes = %d' % (dataframe.shape[1]))

```

```python
dataframe.corr()
```

Features needs to be standardized because there is a difference in magnitude between them. When calculating the principal components other features are going to dominate other features because of the larger magnitude scale and, therefore, larger range of variance.

```python
X = dataframe.loc[:, ['Profile factor','Symmetry','Crossflow','Swirl angle','Flow velocity 1','Flow velocity 2','Flow velocity 3','Flow velocity 4',
'Average flow', 'Speed of Sound 1','Speed of Sound 2','Speed of Sound 3','Speed of Sound 4','Average Speed','Signal Strength 1','Signal Strength 2','Signal Strength 3','Signal Strength 4','Signal Strength 5',
'Signal Strength 6','Signal Strength 7','Signal Strength 8', 'Turbulence 1','Turbulence 2','Turbulence 3','Turbulence 4','Meter performance','Signal Quality 1','Signal Quality 2','Signal Quality 3','Signal Quality 4','Signal Quality 5','Signal Quality 6','Signal Quality 7','Signal Quality 8',
'Gains 1','Gains 2','Gains 3','Gains 4','Gains 5','Gains 6','Gains 7','Gains 8',
'Transit Time 1','Transit Time 2','Transit Time 3','Transit Time 4','Transit Time 5','Transit Time 6','Transit Time 7','Transit Time 8']]
Y = dataframe.loc[:, ["Class"]]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

Proceed with the calculation of the principal components.

```python
from sklearn.decomposition import PCA
PCA = PCA(n_components=2)
components = PCA.fit_transform(X)
PCA.components_
```

Eigenvalues the variance explained by the two principal components and the cumulative explained variance.

```python
cumVar = pd.DataFrame(np.cumsum(PCA.explained_variance_ratio_)*100, 
                      columns=["cumVarPerc"])
expVar = pd.DataFrame(PCA.explained_variance_ratio_*100, columns=["VarPerc"])
pd.concat([expVar, cumVar], axis=1)\
    .rename(index={0: "PC1", 1: "PC2"})
```

The first principal component explains 39.25% of the total variation of the original data, while the second explains 25.98%. Together, the two principal components explain about 65.23% of the total variation.

```python
import seaborn as sns
componentsDf = pd.DataFrame(data = components, columns = ['PC1', 'PC2'])
pcaDf = pd.concat([componentsDf, Y], axis=1)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=pcaDf, x="PC1", y="PC2", hue="Class",palette=['red','green','blue'], legend='full')
```

The scatter plot is used to see the values of the observations with respect to the two principal components and to highlight the different observations by their Classes.

**<font color="red">Observations:</font>** 
It can been seen that there is a clear separation between the different types of Classes. Each Class represents a specific cluster and the distribution can be seen in the graph. 

```python
def biplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley,s=5)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'green', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
 
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()
```

```python
plt.figure(figsize=(20, 20))
biplot(components, np.transpose(PCA.components_), list(dataframe.columns))
```
