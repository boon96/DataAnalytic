```python
# Splitting the attributes into independent and dependent attributes
X = dataframe.iloc[:,dataframe.columns != 'Class']
y = dataframe.Class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=42)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
print(f'model score on training data: {model.score(X_train, y_train)}')
print(f'model score on testing data: {model.score(X_test, y_test)}')
```

```python
# Dropping non-important features
columns_to_drop=['Speed of Sound 2','Signal Strength 2', 'Signal Strength 1','Transit Time 4', 'Speed of Sound 3', 'Gains 4'
,'Speed of Sound 1','Signal Strength 3','Flow velocity 4','Flow velocity 3','Flow velocity 2','Flow velocity 1'
,'Transit Time 5','Transit Time 6','Average flow','Signal Strength 6','Signal Strength 4','Signal Quality 3'
, 'Gains 2','Gains 1','Transit Time 1','Signal Quality 6','Signal Quality 5','Signal Quality 4','Signal Quality 2'
, 'Gains 3','Signal Quality 1', 'Transit Time 2','Turbulence 3','Turbulence 2','Turbulence 1','Transit Time 3','Transit Time 8']
dataframe=dataframe.drop(columns_to_drop,axis=1)

print('Number of instances = %d' % (dataframe.shape[0]))
print('Number of attributes = %d' % (dataframe.shape[1]))
dataframe
```

Features needs to be standardized because there is a difference in magnitude between them. When calculating the principal components other features are going to dominate other features because of the larger magnitude scale and, therefore, larger range of variance.

```python
new_X = dataframe.iloc[:,dataframe.columns != 'Class']
new_y = dataframe.Class

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

Proceed with the calculation of the principal components.

```python
from sklearn.decomposition import PCA
PCA = PCA(n_components=0.95)
components = PCA.fit_transform(X)
PCA.components_
```

Eigenvalues the variance explained by the two principal components and the cumulative explained variance.

```python
cumVar = pd.DataFrame(np.cumsum(PCA.explained_variance_ratio_)*100, 
                      columns=["cumVarPerc"])
expVar = pd.DataFrame(PCA.explained_variance_ratio_*100, columns=["VarPerc"])
pd.concat([expVar, cumVar], axis=1)
```

```python
X_pca = pd.DataFrame(components)
print(X_pca.head())
print("\nSize: ")
print(X_pca.shape)
```

```python
fig = plt.figure(figsize=(10, 8))
sns.heatmap(X_pca.corr(), 
            annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

The first principal component explains 39.25% of the total variation of the original data, while the second explains 25.98%. Together, the two principal components explain about 65.23% of the total variation.

The scatter plot is used to see the values of the observations with respect to the two principal components and to highlight the different observations by their Classes.

**<font color="red">Observations:</font>** 
It can been seen that there is a clear separation between the different types of Classes. Each Class represents a specific cluster and the distribution can be seen in the graph. 

```python
def biplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley,s=5)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'green', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
 
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()
```
